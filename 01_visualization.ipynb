## 1. Quaternion Basis Visualization

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Basis vectors
origin = np.array([0, 0, 0])
vectors = {
    'i': (1, 0, 0),
    'j': (0, 1, 0),
    'k': (0, 0, 1)
}
colors = {'i': 'red', 'j': 'green', 'k': 'blue'}

for name, vec in vectors.items():
    ax.quiver(*origin, *vec, color=colors[name], arrow_length_ratio=0.1, linewidth=2)
    ax.text(*vec, f'${name}$', fontsize=14, color=colors[name])

ax.set_xlim([-1.2, 1.2])
ax.set_ylim([-1.2, 1.2])
ax.set_zlim([-1.2, 1.2])
ax.set_xlabel('i')
ax.set_ylabel('j')
ax.set_zlabel('k')
ax.set_title('Quaternion Basis Vectors')

plt.show()
```
